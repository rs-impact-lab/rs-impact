```python
# CÉLULA 1: Configuração do ambiente e caminhos
import pandas as pd
import os

ARQUIVO_CLIMA = "clima_rs_mensal.csv" 
CAMINHO_CLIMA = os.path.join("..", "data", "processed", "clima", ARQUIVO_CLIMA)

print(f"✅ Ambiente pronto.")
print(f"🔍 Buscando dados climáticos em: {os.path.abspath(CAMINHO_CLIMA)}")
```

```python
# CÉLULA 2: Carregamento dos dados
if os.path.exists(CAMINHO_CLIMA):
    df_clima = pd.read_csv(CAMINHO_CLIMA, sep=';', encoding='latin-1', skiprows=8)
    print(f"📊 Sucesso! Dados carregados: {df_clima.shape[0]} registros.")
else:
    print(f"❌ Arquivo não encontrado em: {os.path.abspath(CAMINHO_CLIMA)}")
```

```python
# CÉLULA 3: Visualização inicial e limpeza de colunas
df_clima = df_clima.dropna(axis=1, how='all')

print("📋 Primeiras linhas dos dados meteorológicos:")
display(df_clima.head())
```

```python
# CÉLULA 4: Padronização de Datas e Tipos
coluna_data = 'Data'

if coluna_data in df_clima.columns:
    df_clima[coluna_data] = pd.to_datetime(df_clima[coluna_data], errors='coerce')
    print(f"📅 Período analisado: {df_clima[coluna_data].min()} até {df_clima[coluna_data].max()}")
else:
    print(f"⚠️ Coluna '{coluna_data}' não encontrada. Verifique as colunas existentes: {list(df_clima.columns)}")
```

```python
# CÉLULA 5: Filtragem de Eventos Extremos (Precipitação)
col_chuva = [c for c in df_clima.columns if 'PRECIPITA' in c.upper()]

if col_chuva:
    c = col_chuva[0]
    
    df_clima[c] = pd.to_numeric(df_clima[c].astype(str).str.replace(',', '.'), errors='coerce')
    
    
    chuvas_intensas = df_clima[df_clima[c] > 50]
    print(f"⛈️ Foram encontrados {len(chuvas_intensas)} dias com precipitação acima de 50mm.")
    display(chuvas_intensas.sort_values(by=c, ascending=False).head())
else:
    print("❌ Coluna de precipitação não encontrada.")
```
