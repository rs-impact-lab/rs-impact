```python
# CÉLULA 1: Configuração do ambiente e definição de caminhos.
import pandas as pd
from pathlib import Path

#NOTE: Defina o arquivo e o ano processado.
ARQUIVO_ALVO = "12-tabelas-dezembro-de-2024_tabela_3.parquet"
ANO = "2024"

RAIZ = Path("..").resolve().parents[0]
CAMINHO_DADOS = RAIZ / "rs-impact" / "data" / "processed" / "caged" / ANO / ARQUIVO_ALVO

print(f"✅ Caminho configurado.")
print(f"🔍 Alvo: {CAMINHO_DADOS}")
```

```python
# CÉLULA 2: Carregamento dos dados para a memória RAM.
if CAMINHO_DADOS.exists():
    df = pd.read_parquet(CAMINHO_DADOS)
    print(f"📊 Sucesso! O arquivo tem {df.shape[0]} linhas e {df.shape[1]} colunas.")
else:
    print(f"❌ Erro: Arquivo não encontrado em {CAMINHO_DADOS}")
```

```python
# CÉLULA 3: Visualização inicial da estrutura dos dados.
df.head(10)
```

```python
# CÉLULA 4: Listagem técnica das colunas.
print("🔍 Colunas encontradas (Normalizadas):")
for i, col in enumerate(df.columns):
    print(f"Coluna {i}: '{col}'")
```

```python
# CÉLULA 5: Diagnóstico de Escopo Econômico (Nacional vs Estadual)

print("📋 Primeiras linhas para entender o conteúdo real:")
display(df.head(10))

# Vamos procurar se existe alguma coluna que faça referência a Estados (UF) ou Regiões
colunas_uf = [col for col in df.columns if 'uf' in col.lower() or 'estado' in col.lower() or 'região' in col.lower()]

if colunas_uf:
    print(f"📍 Colunas de localização encontradas: {colunas_uf}")
    for col in colunas_uf:
        print(f"\nValores únicos na coluna '{col}':")
        print(df[col].unique())
else:
    print("\n⚠️ Nenhuma coluna explícita de 'UF' ou 'Estado' foi encontrada.")
    print("Isso indica que esta tabela específica pode ser um consolidado Nacional ou Setorial puro.")
```

```python
# CÉLULA 6: Verificação de linhas de somatório (Subtotais) - CORRIGIDA
linhas_suspeitas = df[df['município'].str.contains('total|ignorado|subtotal', case=False, na=False)]

if not linhas_suspeitas.empty:
    print("⚠️ Atenção! Foram encontradas linhas de agregação na tabela:")
    display(linhas_suspeitas[['município', 'dezembro2024_sem_ajuste_saldos']])
else:
    print("✅ Limpo! A tabela só contém municípios puros.")
```

```python
# CÉLULA 7: Consolidação estatística para o estado do Rio Grande do Sul
saldo_mes_rs = df['dezembro2024_sem_ajuste_saldos'].sum()
admissoes_mes_rs = df['dezembro2024_sem_ajuste_admissões'].sum()
desligamentos_mes_rs = df['dezembro2024_sem_ajuste_desligamentos'].sum()

acumulado_saldo_ano_rs = df['acumulado_do_ano_2024_com_ajustes_saldos'].sum()

print(f"📊 --- Consolidação RS ({ARQUIVO_ALVO}) ---")
print(f"💼 Admissões no Mês: {admissoes_mes_rs:,.0f}".replace(",", "."))
print(f"📉 Desligamentos no Mês: {desligamentos_mes_rs:,.0f}".replace(",", "."))
print(f"🟩 Saldo Líquido do Mês (Dez/2024): {saldo_mes_rs:,.0f}".replace(",", "."))
print(f"🏆 Saldo Acumulado do Ano (2024): {acumulado_saldo_ano_rs:,.0f}".replace(",", "."))
```

```python
# CÉLULA 8: Varredura de arquivos e construção da série temporal do CAGED
import re

pasta_2024 = CAMINHO_DADOS.parent
arquivos_parquet = sorted(list(pasta_2024.glob("*.parquet")))

dados_serie_temporal = []

print(f"🔄 Iniciando processamento de {len(arquivos_parquet)} arquivos em {pasta_2024.name}...\n")

for caminho_arq in arquivos_parquet:
    nome_fich = caminho_arq.name
    
    # Extrai o número do mês do nome do arquivo (ex: "12-tabelas..." -> "12")
    match_mes = re.search(r'^(\d+)', nome_fich)
    if not match_mes:
        continue
    mes = match_mes.group(1).zfill(2)
    ano_mes = f"{ANO}-{mes}"
    
    try:
        df_mes = pd.read_parquet(caminho_arq)
        
        # Identifica dinamicamente a coluna de saldo daquele mês específico
        coluna_saldo = [col for col in df_mes.columns if 'saldos' in col.lower() and 'acumulado' not in col.lower()]
        
        if coluna_saldo:
            col_alvo = coluna_saldo[0]
            # Filtra garantindo apenas municípios puros (evitando possíveis subtotais)
            df_limpo = df_mes[~df_mes['município'].str.contains('total|ignorado|subtotal', case=False, na=False)]
            
            saldo_consolidado = df_limpo[col_alvo].sum()
            
            dados_serie_temporal.append({
                "ano_mes": ano_mes,
                "saldo_empregos_rs": saldo_consolidado
            })
            print(f"✅ Processado: {ano_mes} | Coluna usada: '{col_alvo}' | Saldo: {saldo_consolidado:,.0f}".replace(",", "."))
        else:
            print(f"⚠️ Aviso: Coluna de saldo não identificada no arquivo {nome_fich}")
            
    except Exception as e:
        print(f"❌ Erro ao processar {nome_fich}: {e}")

# Transforma a lista de dicionários em um DataFrame estruturado
df_caged_cronologico = pd.DataFrame(dados_serie_temporal).sort_values("ano_mes").reset_index(drop=True)

print("\n📈 Série Temporal do CAGED RS para 2024 criada com sucesso!")
display(df_caged_cronologico)
```
