```python
# CÉLULA 1: Configuração do ambiente e definição de caminhos.
# Aqui importamos as bibliotecas necessárias e definimos onde o arquivo está localizado.
# O uso de '..' serve para subir um nível de pasta (de /notebooks para a raiz do projeto).
import pandas as pd
import os
#NOTE: Mude o arquivo alvo aqui.
ARQUIVO_ALVO = "3-tabelas_dezembro-de-2023_Tabela_3.csv" 

CAMINHO_DADOS = os.path.join("..", "data", "raw", "caged", "2023", ARQUIVO_ALVO)

print(f"✅ Caminho configurado.")
print(f"🔍 Olhando em: {os.path.abspath(CAMINHO_DADOS)}")
```

```python
# CÉLULA 2: Carregamento dos dados para a memória RAM.
# Verifica se o arquivo existe antes de tentar ler para evitar erros de execução.
# O encoding 'utf-8-sig' garante a leitura correta de acentos em arquivos CSV vindos do Excel.
if os.path.exists(CAMINHO_DADOS):
    df = pd.read_csv(CAMINHO_DADOS, encoding='utf-8-sig')
    print(f"📊 Sucesso! O arquivo tem {df.shape[0]} linhas e {df.shape[1]} colunas.")
else:
    print(f"❌ Erro: Arquivo não encontrado em {os.path.abspath(CAMINHO_DADOS)}")
```

```python
# CÉLULA 3: Visualização inicial da estrutura dos dados.
# Mostra as 10 primeiras linhas para entendermos o formato das informações e como elas estão distribuídas.
df.head(10)
```

```python
# CÉLULA 4: Listagem técnica das colunas.
# Essencial para identificar os nomes exatos das colunas (UF, Município, Saldos, etc.)
# Serve como guia para configurar os filtros nas próximas etapas de análise.
print("🔍 Colunas encontradas no arquivo:")
for i, col in enumerate(df.columns):
    print(f"Coluna {i}: '{col}'")
```

```python
# CÉLULA 5: Extração e filtragem dos dados de interesse.
# Filtra o DataFrame original para isolar apenas os registros do Rio Grande do Sul (RS).
# Essa célula transforma o dado bruto em dado útil para o projeto rs-impact.

#NOTE: Troque para a coluna desejada.
coluna_desejada = 'UF' 

if coluna_desejada in df.columns:
    df_rs = df[df[coluna_desejada].str.contains('RS', na=False)]
    print(f"📍 Registros do RS: {len(df_rs)}")
    display(df_rs.head())
else:
    print(f"⚠️ A coluna '{coluna_desejada}' não foi encontrada neste arquivo.")
```
